```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# -------------------------
# Data original

# -------------------------
segmento_norte = {
    'Meses': ['Enero', 'Febrero', 'Marzo', 'Abril'],
    'Total_Venta': [100000, 150000, 130000, 250000]
}

df = pd.DataFrame(segmento_norte)
df['Mes_Num'] = range(1, len(df) + 1)

display(df)

# -------------------------
# Modelo de forecast (tendencia lineal)
# -------------------------
coef = np.polyfit(df['Mes_Num'], df['Total_Venta'], 1)
modelo = np.poly1d(coef)

# -------------------------
# Forecast próximos 3 meses
# -------------------------
meses_futuros = [5, 6, 7]
ventas_forecast = modelo(meses_futuros)

df_forecast = pd.DataFrame({
    'Meses': ['Mayo', 'Junio', 'Julio'],
    'Forecast_Venta': ventas_forecast.astype(int)
})

display(df_forecast)

# -------------------------
# Gráfica real + forecast
# -------------------------
plt.plot(df['Mes_Num'], df['Total_Venta'], marker='o', label='Ventas reales')
plt.plot(meses_futuros, ventas_forecast, marker='o', linestyle='--', label='Forecast')

plt.xticks(
    ticks=[1,2,3,4,5,6,7],
    labels=['Enero','Febrero','Marzo','Abril','Mayo','Junio','Julio']
)

plt.xlabel('Meses')
plt.ylabel('Total de ventas')
plt.title('Forecast Ventas Segmento Norte')
plt.legend()
plt.show()
```
